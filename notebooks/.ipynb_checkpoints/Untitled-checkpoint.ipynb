```python
def filtrado(x):
    remove = set()
    na_total = x.isnull().sum()
    for i in na_total.index:
        if (na_total[i]>50000):
            remove.add(i)
    for i in remove:
        x.drop(i, axis=1, inplace=True)
```

```python
filtrado(raw_data)
```
