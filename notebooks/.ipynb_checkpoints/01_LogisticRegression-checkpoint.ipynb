```python
# Importamos las librer√≠as necesarias
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
lc_data = pd.read_csv("../data/loan_clean.csv")
lc_data.drop('Unnamed: 0', inplace=True, axis=1)
```

```python
lc_data
```

```python
# Define the predictors (x) and the target variable (y)

y = lc_data['loan_status']
X = lc_data.drop('loan_status', axis=1)
```

```python
# Split the sample into training and testing, with 25% for testing and the rest for training

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42, stratify=y)
```

```python
# Import package for Logistic Regression
from sklearn.linear_model import LogisticRegression
```

```python
# Instance the regressor
logisticRegr = LogisticRegression()
```

```python
# Scale the train sample. Do this for both test and train?
from sklearn import preprocessing
x_scaled = preprocessing.scale(x_train)
x_scaled
```

```python
# Fit the model. Esto es lo que tarda. Pero bueno este no
logisticRegr.fit(x_train, y_train)
```

```python
def save_models(filename, model):
    with open(filename, 'wb') as file:
        pickle.dump(model, file)
```

```python
import pickle
```

```python
filename = 'logisticRegr.sav'
pickle.dump(logisticRegr, open(filename, 'wb'))
```

```python
# Predict for test set
predictions = logisticRegr.predict(x_test)
```

```python
# Use score method to get accuracy of model
score = logisticRegr.score(x_test, y_test)
print(score)
```

```python
# Show confusion matrix for test sample

from sklearn.metrics import plot_confusion_matrix
plot_confusion_matrix(logisticRegr, x_test, y_test)
```
